```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
```

```python
# First, We'll create our own dataset with 4 data blobs
X,Y = make_blobs(n_samples=500,centers=4,random_state=20)
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
#Now, We import our module
from Kmeans_module import Kmeans
```

```python
kmeans=Kmeans(no_of_clusters=4,max_itr=300)
```

```python
#Training our kmeans model
kmeans.fit(X)
```

```python
#Lets plot the data and clustures to check how well our algorithm performs
plt.scatter(X[:,0],X[:,1], s=7)
plt.scatter(kmeans.centroids[:,0], kmeans.centroids[:,1], marker='*', c='r', s=150)
plt.show()
```
