```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from linear_regression_module import LinearRegression
```

```python
lg=LinearRegression()
```

```python
X = np.load('./data/LinearX.npy')
Y = np.load('./data/LinearY.npy')
# X = X.values
# Y = Y.values
```

```python
X=X.reshape(-1,1)
```

```python
X=lg.normalize(X)
```

```python
X=lg.add_ones(X)
```

```python
X.shape
```

```python
Y=Y.reshape(-1,1)
```

```python
Y.shape
```

```python
plt.scatter(X[:,1],Y)
```

```python
theta,cost_list = lg.gradient_descent(X,Y,0.05,200)
```

```python
theta
```

```python
plt.plot(cost_list)
plt.show()
```

```python
plt.scatter(X[:,1],Y)
Y_ = lg.hypothesis(theta,X)
plt.plot(X[:,1],Y_,color='red')
plt.show()
```
