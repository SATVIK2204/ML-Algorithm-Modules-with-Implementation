```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from logistic_regression_module import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
X = pd.read_csv('./data/logisticX.csv')
Y = pd.read_csv('./data/logisticY.csv')
X = X.values
Y = Y.values
```

```python
X=lg.normalize(X)
```

```python
X=lg.add_ones(X)
```

```python
X.shape,Y.shape
```

```python
plt.style.use("seaborn")
plt.scatter(X[:,1],X[:,2],c=Y[:,0],cmap=plt.cm.Accent)
plt.show()
```

```python
theta,cost_list = lg.gradient_descent(X,Y,0.01,200)
```

```python
theta
```

```python
plt.plot(cost_list)
plt.show()
```

```python
def visualiseBoundary():
    plt.style.use("seaborn")
    plt.scatter(X[:,1],X[:,2],c=Y[:,0],cmap=plt.cm.Accent)
    x1 = np.linspace(-3,+3,5)
    x2 = -(theta[0] + theta[1]*x1)/theta[2]
    plt.plot(x1,x2)
    plt.show()
def predict(theta,X):
    Y_pred = lg.hypothesis(theta,X)
    outputs = np.zeros((Y_pred.shape))
    outputs[Y_pred>0.5]=1
    return outputs
```

```python
Y_pred = predict(theta,X)
```

```python
Y_pred.shape
```

```python
accuary_on_train_set = np.sum(Y_pred==Y)/X.shape[0]
```
